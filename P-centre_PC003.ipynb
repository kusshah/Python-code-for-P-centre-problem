```python
######### Gurobi Code for P Centre Problem ---PC003 with time limit


from gurobipy import*

#open the file

file=open("PC003.dat")

# split values from lines when lenghth is greater than 0.
for L in file:
    a=L.rstrip()
    if len(a)>0:
        a=L.split()
        if a[0]=='N:':
            P=a[1]
        elif a[0]=='P:':
            c_r=a[1]
    else:
        continue

d={}
b=0
e=0
l=int(P)
q=int(P)
p=int(c_r)

#open file
file=open("PC003.dat")

# split values from lines when lenghth is greater than 3

for L in file:
    L=L.split()
    if len(L)>3:
        for h in L:
            d[b,e]=h
            e=e+1
        e=0
        b=b+1 
    else:
        continue
        
#define model
m=Model("P_CENTRE")

#set Decision variables
x={}
y={}
MD=m.addVar(vtype=GRB.CONTINUOUS,name='MAX_DISTANCE')
for j in range(q):
    y[j]=m.addVar(vtype=GRB.BINARY,name='binary')
    for i in range(l):
        x[(i,j)]=m.addVar(vtype=GRB.BINARY,name='Binary')

#set objective
m.setObjective(MD, GRB.MINIMIZE)




#set constraints

#sum of total distance should be less than decsion variable

for i in range(l):
    for j in range(q):
        m.addConstr(x[(i,j)]*d[(i,j)] <= MD)
        
# each non-centre point is assigned to only one centre

for i in range(l):
    m.addConstr((quicksum(x[(i,j)] for j in range(q))) ==1,)

#total number of centres are equal to p    

m.addConstr(quicksum(y[j] for j in range(q))==p)

# each centre point is assigned to non-centre point if that point is defined as centre 

for j in range(q):
    for i in range(l):
        m.addConstr(x[(i,j)] <= y[j])
        
#  set time limit
m.params.timelimit=600

# optimize model

m.optimize()

#Print objective value

print("The maximum distance between the non-center to its center is %e" % m.objVal, )

#print centre and non-centre point and which non-centre point has been assigned to which centre point

for j in range(q):
    if y[j].X==1:
        print("The point", j+1, "is a center")
    elif y[j].X==0:
        print("The point", j+1,"is a non center")
        
for j in range(q):
    for i in range(l):
        if x[(i,j)].X==1:
            print("The non center point", i+1, "is assigned to center point", j+1)




#thank you    

        

```
